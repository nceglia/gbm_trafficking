```python
import trafficking as trf
import scanpy as sc
```

```python
adata = sc.read("../data/objects/GBM_TCR_POS_TCELLS.h5ad")
```

```python
result, data, phenotypes = trf.run_inference(
    adata, "PBMC", "TP", lineage="CD8", temporal=True
)
```

```python
# Visualize
trf.plot_elbo(result)
trf.plot_posterior(result, "PBMC", "TP")
trf.plot_patient_heterogeneity(result, "PBMC", "TP")
trf.plot_posterior_intervals(result)
trf.plot_kappa(result)
trf.plot_credible(result, threshold=0.1)

# Credible transitions table
df = trf.credible_transitions(result, threshold=0.1, ci=0.9)
print(df[df["credible"]])

# Directional comparison
comp = trf.compare_directions(adata, "PBMC", "TP", lineage="CD8")
trf.plot_asymmetry(comp)
```
